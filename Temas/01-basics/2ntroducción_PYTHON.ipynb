### Cortes de Listas **(RANGOS)** 

Como habíamos visto anteriormente, podremos consultar datos dentro de nuestras lista.
Pero, ¿qué pasa si queremos consultar un rango de datos?

```python
familia_edades1 = ['Eduardo', 14, 'Mario', 25 , 'Karla' , 45 , 'Natalia', 46 , 'Guadalupe', 89 ]
```

Para poder consultar un rango de datos ocuparemos la siguiente estructura:

` print(nombredelalista[INICIO : FINAL]) `

*Dato:*

*AL PONER EL RANGO FINAL SE LE RESTARÁ UN -1 PARA PODER OBETENER EL HASTA DONDE ESTARÁ IMPRIMIENTO ) *

```python
print(familia_edades1[2:4])
```

Como podemos observar el elemento 4 ` Karla ` no se presenta, esto debido a que el corte lo maneja como Final -1 y al momento de consultar los datos se presentaran los datos puestos en el rango, menos el último.

**AGREGAR ELEMENTOS A LA LISTA**

Para poder agregar un nuevo elemento a tu lista tendrás que seguir la siguiente estructura:

x= nuevo elemento

y= nuevo elemento

` nommbredelalista + [ x ] `

` nommbredelalista + [ x , y ] `


```python
fam3 = ['Eduardo', 14, 'Mario', 25 , 'Karla' , 45 , 'Natalia', 46 , 'Guadalupe', 89 ]
```

```python
fam_nuevo_elemento = fam3 + ["Alberto" , 30]
```

```python
print(fam_nuevo_elemento)
```

**REMOVER O ELIMINAR ELEMENTO**

Para poder remover un elemento de una lista, se seguirá el siguiente formato:

x= elemento a eliminar

del(nombredelalista[x])

```python
fam3 = ['Eduardo', 14, 'Mario', 25 , 'Karla' , 45 , 'Natalia', 46 , 'Guadalupe', 89 ]
```

```python
del (fam3[2])
```

```python
print(fam3)
```

Al momento de eliminar un elemento, los posiciones o los indices se organizan de nuevo, por lo que, si vuelvo a querer eliminar el elemento que se encuentra en la posición 2 estaré eliminando el elemento *25*

```python
del (fam3[2])
```
